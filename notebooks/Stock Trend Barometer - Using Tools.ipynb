# Stock Trend Barometer

### Extract S&P500 tickers from Wikipedia using pandas, use these to collect historic prices from Yahoo Finance and calculate multiple Technical Indicators over various time frames and determine the strength of a trend by aggregating these indicators

### Imports

```python
import barometer as bar
```

### Initialise market

```python
mkt = bar.DataSetYahoo('2018-08-10', '2020-09-28')
```

### Extract market data, calculate indicators and trend strength

```python
mkt.prepyahoo().createbarometer()
```

### Express the Top 20 results in a horizontal barchart

```python
mkt.trendbarchart(mkt.barometer, 20)
```

### And the Bottom 20

```python
mkt.trendbarchart(mkt.barometer, 20, top=False)
```

### Prepare chart data

```python
mkt.prepdata()
```

### Normalized dataframes covering the past 1y, 6m and 3m

```python
norm_3m = mkt.normalise(mkt.chart_data, 60) 
norm_6m = mkt.normalise(mkt.chart_data, 120) 
norm_1y = mkt.normalise(mkt.chart_data, 250)
```

### 3 Month Returns of top and bottom 5 trending markets

```python
mkt.returnsgraph(norm_3m)
```

### 6 Month Returns of top and bottom 5 trending markets

```python
mkt.returnsgraph(norm_6m)
```

### 1 Year Returns of top and bottom 5 trending markets

```python
mkt.returnsgraph(norm_1y)
```

### Prepare chart data

```python
mkt.prepdata(mkts=20)
```

### 3m Returns for each of the top and bottom 20 trending markets - normalised

```python
mkt.marketchart(days=60)
```

### 3m Returns for each of the top and bottom 20 trending markets - non-normalised

```python
mkt.marketchart(days=60, norm=False)
```

### 1y Returns for each of the top and bottom 20 trending markets - normalised

```python
mkt.marketchart(days=250)
```

### 1y Returns for each of the top and bottom 20 trending markets - non-normalised

```python
mkt.marketchart(days=250, norm=False)
```
